Nesse notebook vamos analisar os dados anotados e sua distribuição, para então prosseguir com o treinamento do modelo.

```python
import pandas as pd

caminho_do_arquivo = '/Users/giossaurus/Developer/leia_tcc/data/processed/seed_set_anotado.csv'

try:
    df_anotado = pd.read_csv(caminho_do_arquivo, quotechar='"', escapechar='\\')

    print("--- INFORMAÇÕES DO SEU DATASET ANOTADO ---")
    
    print("\n[1] Informações Gerais e Nomes das Colunas:")
    df_anotado.info()
    
    print("\n[DEBUG] Primeiros valores de intent_choice:")
    print(df_anotado['intent_choice'].head(10))
    
    coluna_de_anotacao = 'intent_choice'

    if coluna_de_anotacao in df_anotado.columns:
        print(f"\n[2] Distribuição dos Rótulos na Coluna '{coluna_de_anotacao}':")
        print(df_anotado[coluna_de_anotacao].value_counts())
        print(df_anotado['intent_choice'].head(10))
        
        print(f"\n[3] Total de anotações: {len(df_anotado)}")
        print("\nValores únicos:")
        print(df_anotado['intent_choice'].unique())

        print("\nDistribuição atual dos rótulos:")
        print(df_anotado['intent_choice'].value_counts())
    else:
        print(f"\n[ERRO] A coluna '{coluna_de_anotacao}' não foi encontrada.")
        print("Por favor, verifique o nome da coluna de anotação no output acima e ajuste a variável 'coluna_de_anotacao' no script.")

except FileNotFoundError:
    print(f"--- ERRO ---")
    print(f"O arquivo não foi encontrado no caminho especificado: '{caminho_do_arquivo}'")
    print("Por favor, verifique se o caminho está correto e tente novamente.")

```
